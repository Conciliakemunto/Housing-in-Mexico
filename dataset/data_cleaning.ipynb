```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python

df1 = pd.read_csv('C:/Users/conci/Desktop/housing-mexico-project/dataset/mexico-dataset-1.csv')
df2 = pd.read_csv('C:/Users/conci/Desktop/housing-mexico-project/dataset/mexico-dataset-2.csv')
df3 = pd.read_csv('C:/Users/conci/Desktop/housing-mexico-project/dataset/mexico-dataset-3.csv')
```

```python
#inspecting the data
df1.info()
df1.head()
```

```python
#dropping rows null values 

df1.dropna(inplace = True)
df1.info()
```

```python
#Removing $ and commas in price_usd column and converting it to float
df1['price_usd'] = df1['price_usd'].replace(r'[\$,]', '', regex=True).astype(float)
df1.head()
```

```python
#cleaning df2
df2.info()
```

```python
#drop all NaN rows
df2.dropna(inplace= True)
df2.info()
```

```python
#create new column "price_used" from price_mxm(mexican pesos) and drop price_mxn column
#data was collected in 2014, a dollar cost 19 pesos
df2['price_usd'] = (df2['price_mxn'] / 19).round(2)
df2.drop(columns = ['price_mxn'], inplace = True)
```

```python
#cleaning df3
df3.info()
```

```python
df2.head()
```

```python
df3.head()
```

```python
#splitting lat-lon column into two columns 
#lat and lon and changing their data type from object to float
df3[['lat','lon']] = df3['lat-lon'].str.split(',', expand = True).astype(float)
df3.head()
```

```python
#create state column from place_with_parent_names column then drop place_with_parent_names column
df3['state'] = df3['place_with_parent_names'].str.split('|', expand = True)[2]
df3.drop(columns = ['place_with_parent_names'], inplace = True)
df3.head()
```

```python
#drop lat-lon column
df3.drop(columns = ['lat-lon'], inplace = True)
```

```python
#drop all NaN rows
df3.dropna(inplace= True)
```

```python
#combining the three dataframes
df = pd.concat([df1,df2,df3])
print(df.shape)
df.info()
```

```python
df.to_csv('mexico_real_estate_clean.csv',index=False)
```
